```python
# Dealing with NaN and
review_df.dropna(axis=0,inplace=True)
```

```python
# Text cleaning

def text_cleaning(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    text = ' '.join(text.split()) # remove extra spaces
    text = ' '.join([contraction_mapping[t] if t in contraction_mapping else t for t in text.split(" ")])

    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))

    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    return " ".join(words)
```

```python
review_df["Cleaned_text"] = review_df["Text"].apply(text_cleaning)
```

```python
review_df["Cleaned_Summary"] = review_df["Summary"].apply(text_cleaning)
```

```python
from transformers import AutoTokenizer

# Example: T5 (you can use 'facebook/bart-base' instead)
model_name = "t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Tokenize the input texts
input_encodings = tokenizer(
    review_df["Cleaned_text"].tolist(),
    max_length=512,
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)

# Tokenize the target summaries
target_encodings = tokenizer(
    review_df["Cleaned_Summary"].tolist(),
    max_length=128,
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)
```

Convert data into pytorch dataset object

```python
# Convert dataset into pytorch dataset object
import torch
from torch.utils.data import Dataset

class SummarizationDataset(Dataset):
    def __init__(self, inputs, targets):
        self.input_ids = inputs["input_ids"]
        self.attention_mask = inputs["attention_mask"]
        self.labels = targets["input_ids"]

        # Set padding token to -100 for loss ignoring
        self.labels[self.labels == tokenizer.pad_token_id] = -100

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {
            "input_ids": self.input_ids[idx],
            "attention_mask": self.attention_mask[idx],
            "labels": self.labels[idx]
        }

# Create dataset
dataset = SummarizationDataset(input_encodings, target_encodings)
```

```python
# Model Training / Tuning

from transformers import AutoModelForSeq2SeqLM, Trainer, TrainingArguments
# model_name = "facebook/bart-large-cnn"
model_name = "t5-base"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)


# Disable wandb by turning off reporting
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=10, #defines how many times your model will see all the training data during training.
    per_device_train_batch_size=2, #Each training step processes 4 samples on each device (GPU or CPU).
    logging_strategy="steps",     # log loss every N steps
    logging_steps=1,              # print every step
    logging_dir="./logs",
    report_to="none",       # <- THIS disables wandb
    save_strategy="no"      # optional: disables model checkpoint saving
)

# By default, the loss is computed using the model’s internal loss function, typically:Cross Entropy Loss

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset
)

# Train the model
trainer.train()
```

```python
# Model Inference
    
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

def generate_summaries(model, tokenizer, texts, max_input_length=512, max_output_length=50, num_beams=4, print_output=True):
    """
    Generates summaries for a list of input texts using the given model and tokenizer.
    Ensures everything runs on the same device (CPU or GPU).
    """

    # Determine device (GPU if available)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    text = text_cleaning(texts)

    # Tokenize and move to device
    inputs = tokenizer(
        texts,
        max_length=max_input_length,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    inputs = {k: v.to(device) for k, v in inputs.items()}  # move all tensors to same device

    # Generate summaries
    model.eval()
    with torch.no_grad():
        generated_ids = model.generate(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_length=max_output_length,
            num_beams=num_beams,
            early_stopping=True
        )

    # Decode summaries (move output back to CPU first)
    summaries = tokenizer.batch_decode(generated_ids.cpu(), skip_special_tokens=True)

    if print_output:
        for i, summary in enumerate(summaries):
            print(f"\nOriginal Text: {texts[i]}\nGenerated Summary: {summary}")

    return summaries
```

Q1) What is Cross Entropy?
-------------------------------------------
Cross-entropy measures the difference between two probability distributions:

The true distribution (what should happen — your labels)

The predicted distribution (what the model thinks will happen)

It answers:

"How surprised is the model by the actual result?"

Less surprise = lower cross-entropy = better predictions
In the context of NLP / summarization:
For every token the model generates (e.g., a word or subword), it predicts a probability distribution over the entire vocabulary.

Example:
Vocabulary = [dog, cat, bird, food]

Actual next word: "cat" → [0, 1, 0, 0] (true one-hot distribution)

Model predicts: [0.2, 0.5, 0.2, 0.1]

Cross-entropy measures how far [0.2, 0.5, 0.2, 0.1] is from [0, 1, 0, 0].

Q3) What is T5 (Text-to-Text Transfer Transformer)
-------------------------------------------
t%is a language model developed by Google that treats every NLP task as a text-to-text problem.
It uses a Transformer-based encoder-decoder architecture, similar to models like BART and PEGASUS.
T5 is pretrained on a large corpus (C4) using a masked span prediction objective.
It can be used for tasks like summarization, translation, classification, and question answering with just text prompts.
T5 comes in multiple sizes (t5-small, t5-base, t5-large, etc.), making it flexible for both speed and accuracy.
